```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x=tips["sex"],y=tips["total_bill"],data=tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x="day",y=tips["total_bill"],data=tips,hue="smoker")
```

```python
sns.boxplot(x="day",y=tips["total_bill"],data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,split=True,hue="sex")
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
# drawback: for large dataset, it doesn't show very clearly
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,kind="bar")
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,kind="violin")
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,kind="strip")
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,kind="box")
```
