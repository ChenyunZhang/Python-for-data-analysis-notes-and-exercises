```python
import matplotlib.pyplot as plt
```

%matplotlib inilne

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,"red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("String Title")
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--');
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.6,0.5,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel("X_label_1")
axes1.set_ylabel("Y_label_1")
axes1.set_title("Axies 1 Title")

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,"r")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("title")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x,y,"b")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("title")
    
fig
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

for ax in axes:
    ax.plot(x,y,"g")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("title")

fig
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X square")
ax.plot(x,y*2,label="Y 2times")
ax.legend(loc=0)
```
